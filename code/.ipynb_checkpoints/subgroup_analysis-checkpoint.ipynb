# Import data

```python
PPGR_meal_merged = pd.read_csv('../../data/PPGR_meal_merged_cont2.csv',index_col=0) #Merged data (meal + CGM)
OTU_data = pd.read_csv('../../data/T2D_KBSMC_otu_norm.csv',index_col=0)             #16s microbiome data (taxonomic mapped read count profile)
med_data = pd.read_excel('../../data/CGM Nutrition_CRF_Medication.xlsx')            #Medication usage
clinical_data = pd.read_excel('../../data/CGM Nutrition_CRF_20230808_추가Lab.xlsx') #Clinicodemograhic data
```

```python
# Calculating PPGR #
CGM_post60_col =  [f"p_{t:03d}" for t in range(0,61,5)]
CGM_post120_col = [f"p_{t:03d}" for t in range(0,121,5)]
CGM_post240_col = [f"p_{t:03d}" for t in range(0,241,5)]

PPGR_meal_merged[CGM_post240_col] = PPGR_meal_merged.loc[:,CGM_post240_col].interpolate(axis=1) # Fill the Nan with interpolated value

G0 = PPGR_meal_merged['p_000']
# 2h PPGR #
iAUC_2h = [0]*len(PPGR_meal_merged)
for i in range(1,25):
    Gi = PPGR_meal_merged[CGM_post120_col[i]]
    Gi_1 = PPGR_meal_merged[CGM_post120_col[i-1]]
    Si = ((Gi-G0)+(Gi_1-G0))/2
    iAUC_2h+=Si*5

# 4h PPGR #
iAUC_4h = [0]*len(PPGR_meal_merged)
for i in range(1,49):
    Gi = PPGR_meal_merged[CGM_post240_col[i]]
    Gi_1 = PPGR_meal_merged[CGM_post240_col[i-1]]
    Si = ((Gi-G0)+(Gi_1-G0))/2
    iAUC_4h+=Si*5
    
PPGR_meal_merged['PPGR_u2'] = iAUC_2h/60
PPGR_meal_merged['PPGR_u4'] = iAUC_4h/60
```

```python
PPGR_meal_merged_filt = PPGR_meal_merged.dropna(subset=['meal_m1'],axis=0)
PPGR_meal_merged_filt = PPGR_meal_merged_filt[(PPGR_meal_merged['Energy(kcal)']<2000) & (PPGR_meal_merged['Carb(g)']<250)]
PPGR_meal_merged_filt['Carb_root'] = PPGR_meal_merged_filt['Carb(g)']**(1/2)
PPGR_meal_merged_filt['Carb_pro'] = PPGR_meal_merged_filt['Carb(g)']*4 / (PPGR_meal_merged_filt['Carb(g)']*4+PPGR_meal_merged_filt['Protein(g)']*4+PPGR_meal_merged_filt['Fat(g)']*9) * 100
PPGR_meal_merged_filt['Protein_pro'] = PPGR_meal_merged_filt['Protein(g)']*4 / (PPGR_meal_merged_filt['Carb(g)']*4+PPGR_meal_merged_filt['Protein(g)']*4+PPGR_meal_merged_filt['Fat(g)']*9) * 100
PPGR_meal_merged_filt['Fat_pro'] = PPGR_meal_merged_filt['Fat(g)']*9 / (PPGR_meal_merged_filt['Carb(g)']*4+PPGR_meal_merged_filt['Protein(g)']*4+PPGR_meal_merged_filt['Fat(g)']*9) * 100

OTU_data = OTU_data.loc[:,list(np.sum(OTU_data>0) > len(OTU_data)*0.3)]
microbiome_col = list(OTU_data.columns)
OTU_data_bc = pd.DataFrame(stats.boxcox(OTU_data,lmbda=0.25),columns=OTU_data.columns,index=OTU_data.index)

# Features
CGM_pre30_col = [f"m_{t:03d}" for t in range(30,0,-5)]+['p_000']
futureCGM_120 = [f"p_{t:03d}" for t in range(5,121,5)]
meal_composition = ['Carb(g)','Protein(g)','Fat(g)','Fiber(g)']
meal_context = ['meal_m1','prot_b6h','Time']
cli_col = ['Age','DM_Duration','BMI','HbA1c','SBP','DBP','HDL','LDL','AST(IU/L)','ALT(IU/L)'] 
med_col = ['Basal_Ins_Dose_Unit','MFM_Dose_Total','SU_Dose_Total','DPP4i_Dose_Total','SGLT2i_Dose_Total'] 

# Target features (2h PPGR & 4h PPGR)
PPGR = 'PPGR_u2'
PPGR_4 = 'PPGR_u4'

# Final dataframe
PPGR_data =  PPGR_meal_merged_filt.dropna(subset=['PPGR_u2','PPGR_u4']+CGM_pre30_col,axis=0)
full_data = pd.merge(PPGR_data[['ID','Dname',PPGR,PPGR_4,'Carb(g)','Meal']+CGM_pre30_col+CGM_post240_col+meal_composition+meal_context],OTU_data_bc[microbiome_col],left_on='ID',right_index=True)
full_data = full_data.T.drop_duplicates().T
full_data = pd.merge(full_data,med_data,how='left',on='ID')
full_data = pd.merge(full_data,clinical_data,how='left',on='ID')
full_data[PPGR] = full_data[PPGR].astype(float)
full_data[PPGR_4] = full_data[PPGR_4].astype(float)

samples = list(full_data['ID'].unique())
full_data
```

# Carbohydrate sensitivity

```python
samples = list(full_data['ID'].unique())
carb_sensitivity = []
carb_sensitivity_p = []
for sample in samples:
    data = full_data[full_data['ID']==sample]
    r,p = stats.pearsonr(data['Carb(g)'],data[PPGR])
    carb_sensitivity.append(r)
    carb_sensitivity_p.append(p)
mean_carb_sen = np.mean(carb_sensitivity)

samples = list(evaluation_data['ID'].unique())
carb_sensitivity_eval = []
carb_sensitivity_p_eval = []
for sample in samples:
    data = evaluation_data[evaluation_data['ID']==sample]
    r,p = stats.pearsonr(data['Carb(g)'],data[PPGR])
    carb_sensitivity_eval.append(r)
    carb_sensitivity_p_eval.append(p)
```

```python
def plot_CGM2(data,ax):
    high = data[data['carb_sen']=='high responder']
    low = data[data['carb_sen']=='low responder']

    ax.set_ylim(80,330)
    CGM_post120_col_skip = [col for i,col in enumerate(CGM_post120_col) if i%4==0]
    
    high_label='High'
    low_label='Low'
    
    ax.errorbar(x=CGM_post120_col_skip,y=high[CGM_post120_col_skip].mean(axis=0), yerr=high[CGM_post120_col_skip].std(axis=0),marker='o',
                markersize=6, color='coral',markeredgecolor='black',alpha=0.7,lw=1.5,label=high_label,capsize=3,capthick=1)

    ax.errorbar(x=CGM_post120_col_skip,y=low[CGM_post120_col_skip].mean(axis=0), yerr=low[CGM_post120_col_skip].std(axis=0),marker='o',
                markersize=6, color='#BFBFBF',markeredgecolor='black',alpha=0.7,lw=1.5,label=low_label, capsize=3,capthick=1)

    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.set_xticks(['p_000','p_020','p_040','p_060','p_080','p_100','p_120'])
    ax.set_xticklabels(['0','20','40','60','80','100','120'])
    
    # 각 time point의 그룹간 비교 (t-test)
    p_list = []
    for col in CGM_post120_col_skip:
        res = stats.ttest_ind(high[col].astype(float),low[col].astype(float))
        if 0.01<res.pvalue<=0.05:
            p_list.append('*')
        elif 0.001<res.pvalue<=0.01:
            p_list.append('**')
        elif res.pvalue<=0.001:
            p_list.append('***')
        else:
            p_list.append('')
    for i,col in enumerate(CGM_post120_col_skip):
        ax.text(x = col,y=low[col].mean()+60,s=p_list[i],fontsize=10,horizontalalignment='center')
        
    ax.legend(ncols=2,loc='upper left')
    ax.set_xlabel('Time(min)',y=-0.04)
    ax.set_ylabel('Sensor glucose (mg/dL)',x=0.09)

full_data2 = pd.merge(full_data,carb_sensitivity_df,how='left',on='ID')
carb_0_100 = full_data2[(full_data2['Carb(g)']<100) & (full_data2['Carb(g)']>=0)]
carb_100 = full_data2[full_data2['Carb(g)']>=100]

fig, ax = plt.subplots(figsize=(4, 3))
plot_CGM2(carb_0_100,ax)
plt.savefig('./figure/glucose_lineplot_0_100.jpeg',dpi=300, bbox_inches='tight')
fig, ax = plt.subplots(figsize=(4, 3))
plot_CGM2(carb_100,ax)
plt.savefig('./figure/glucose_lineplot_100.jpeg',dpi=300, bbox_inches='tight')
```

```python
def collapse_otu(df, level):
    OTU_ = pd.DataFrame()
    OTU_list=df.columns
    level_list = list(set([';'.join(OTU.split(';')[0:level]) for OTU in OTU_list]))
    for i in range(len(level_list)):
        the_otu = [OTU for OTU in df.columns if level_list[i] in OTU]
        OTU_[level_list[i]] = df[the_otu].sum(axis=1)
        
    return OTU_
```

```python
genus_species_col = ['\n'.join(col.split(';')[-2:]) for col in microbiome_col]
OTU_data.columns = genus_species_col
tmp = pd.merge(carb_sensitivity_df[['ID','carb_sen']],OTU_data,left_on='ID',right_index=True)

high = tmp.loc[tmp['carb_sen']=='high responder']
low = tmp.loc[tmp['carb_sen']=='low responder']
high_full = pd.merge(high,full_data[['ID',PPGR]+meal_composition],on='ID',how='inner')
low_full = pd.merge(low,full_data[['ID',PPGR]+meal_composition],on='ID',how='inner')

def causal_inference(data):
    causal_inference_df = pd.DataFrame(columns=['X','M','b2','b3','p2','p3'])
    for nutrient in meal_composition:
        for species in genus_species_col:
            scaler = StandardScaler()
    
            data[species] = scaler.fit_transform(data[species].astype(float).values.reshape(-1, 1))
            data[nutrient] = scaler.fit_transform(data[nutrient].astype(float).values.reshape(-1, 1))
    
            mediation_df = mediation_analysis(data=data, x=nutrient, m=species,
                                        y='PPGR_u2', covar=None, alpha=0.05, n_boot=500, seed=2, return_dist=False, logreg_kwargs=None)
            
            p2 = mediation_df.loc[0,'pval']
            b2 = mediation_df.loc[0,'coef']
            p3 = mediation_df.loc[1,'pval']
            b3 = mediation_df.loc[1,'coef']
            p_total = mediation_df.loc[2,'pval']
            p4 = mediation_df.loc[4,'pval']
            
            if  (p2<0.05) and (p3<0.05) and (p4<0.05) and (p_total<0.05):
                res = {'b2':b2,'b3':b3,'p2':p2,'p3':p3}
                res['X'] = nutrient
                res['M'] = species
                causal_inference_df = pd.concat([causal_inference_df,pd.DataFrame(res,index=[0])], ignore_index=True)
                
    return causal_inference_df

causal_inference_df_high = causal_inference(high_full)
causal_inference_df_low = causal_inference(low_full)
```

```python
G = nx.DiGraph()
Mediators = list(causal_inference_df_high['M'].unique())
for _, row in causal_inference_df_high.iterrows():
    # X -> M edge
    G.add_edge(row['X'], row['M'], p_value=row['p2'], color='red' if row['b2'] > 0 else 'blue', alpha=0.5, weight=-np.log10(row['p2']))
    # M -> Y edge
    G.add_edge(row['M'], 'PPGR',  p_value=row['p3'], color='red' if row['b3'] > 0 else 'blue', alpha=0.5,  weight=-np.log10(row['p3']))

pos = {'Carb(g)':[-1, 1], 'Protein(g)':[-1, -1], 'Fat(g)':[-1, -1], 'Fiber(g)':[-1, -3], 
       'PPGR':[1, 0]
      }
for i,m in enumerate(Mediators):
    pos[m] = [0, 1-i]

edges = G.edges(data=True)
colors = [G[u][v]['color'] for u,v,_ in edges]
weights = [np.abs(G[u][v]['weight']) for u,v,_ in edges]

plt.figure(figsize=(8, 9))
plt.ylim(-6,6)
plt.xlim(-1.5, 1.5)

nx.draw_networkx_nodes(G, pos,  node_size=4000,alpha=0)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, arrowstyle='-|>', arrowsize=15, edge_color=colors, width=weights, 
                        min_source_margin=20,min_target_margin=40)

ax = plt.gca()

# Apply alpha to the arrows manually
for i, (u, v, d) in enumerate(G.edges(data=True)):
    arrow = ax.patches[i]
    arrow.set_alpha(0.4)
plt.axis('off')
plt.tight_layout()
#plt.savefig('./figure/microbiome_causal_inference_high.jpeg',dpi=300)
plt.show()
```
