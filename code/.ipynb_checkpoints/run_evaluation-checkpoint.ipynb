```python
import pandas as pd
import import_ipynb
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras import Model, models, layers
from tensorflow.keras.losses import MeanAbsoluteError, MeanSquaredError
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances
from sklearn.cluster import AgglomerativeClustering
import sklearn
from models import *
from sklearn.svm import SVR
from tensorflow.keras.callbacks import *
import ast
import umap.umap_ as umap
from collections import Counter
from PyComplexHeatmap import *

def feature_split(data):
    x_cgm = np.array(data[CGM_pre30_col]).astype(float)
    x_meal_composition = data[meal_composition].astype(float)
    x_meal_context = data[meal_context].astype(float)
    x_cli = data[cli_col].astype(float)
    x_med = data[med_col].astype(float)
    x_microbiome = data[microbiome_col].astype(float)
    
    return x_cgm, x_meal_composition, x_meal_context, x_cli, x_med ,x_microbiome

mae = MeanAbsoluteError()
mse = MeanSquaredError()
```

# Load data

```python
PPGR_meal_merged = pd.read_csv('../../data/PPGR_meal_merged_cont2.csv',index_col=0) #Merged data (meal + CGM)
OTU_data = pd.read_csv('../../data/T2D_KBSMC_otu_norm.csv',index_col=0)             #16s microbiome data (taxonomic mapped read count profile)
med_data = pd.read_excel('../../data/CGM Nutrition_CRF_Medication.xlsx')            #Medication usage
clinical_data = pd.read_excel('../../data/CGM Nutrition_CRF_20230808_추가Lab.xlsx') #Clinicodemograhic data
```

```python
# Calculating PPGR #
CGM_post60_col =  [f"p_{t:03d}" for t in range(0,61,5)]
CGM_post120_col = [f"p_{t:03d}" for t in range(0,121,5)]
CGM_post240_col = [f"p_{t:03d}" for t in range(0,241,5)]

PPGR_meal_merged[CGM_post240_col] = PPGR_meal_merged.loc[:,CGM_post240_col].interpolate(axis=1) # Fill the Nan with interpolated value

G0 = PPGR_meal_merged['p_000']
# 2h PPGR #
iAUC_2h = [0]*len(PPGR_meal_merged)
for i in range(1,25):
    Gi = PPGR_meal_merged[CGM_post120_col[i]]
    Gi_1 = PPGR_meal_merged[CGM_post120_col[i-1]]
    Si = ((Gi-G0)+(Gi_1-G0))/2
    iAUC_2h+=Si*5

# 4h PPGR #
iAUC_4h = [0]*len(PPGR_meal_merged)
for i in range(1,49):
    Gi = PPGR_meal_merged[CGM_post240_col[i]]
    Gi_1 = PPGR_meal_merged[CGM_post240_col[i-1]]
    Si = ((Gi-G0)+(Gi_1-G0))/2
    iAUC_4h+=Si*5
    
PPGR_meal_merged['PPGR_u2'] = iAUC_2h/60
PPGR_meal_merged['PPGR_u4'] = iAUC_4h/60
```

```python
PPGR_meal_merged_filt = PPGR_meal_merged.dropna(subset=['meal_m1'],axis=0)
PPGR_meal_merged_filt = PPGR_meal_merged_filt[(PPGR_meal_merged['Energy(kcal)']<2000) & (PPGR_meal_merged['Carb(g)']<250)]
PPGR_meal_merged_filt['Carb_root'] = PPGR_meal_merged_filt['Carb(g)']**(1/2)
PPGR_meal_merged_filt['Carb_pro'] = PPGR_meal_merged_filt['Carb(g)']*4 / (PPGR_meal_merged_filt['Carb(g)']*4+PPGR_meal_merged_filt['Protein(g)']*4+PPGR_meal_merged_filt['Fat(g)']*9) * 100
PPGR_meal_merged_filt['Protein_pro'] = PPGR_meal_merged_filt['Protein(g)']*4 / (PPGR_meal_merged_filt['Carb(g)']*4+PPGR_meal_merged_filt['Protein(g)']*4+PPGR_meal_merged_filt['Fat(g)']*9) * 100
PPGR_meal_merged_filt['Fat_pro'] = PPGR_meal_merged_filt['Fat(g)']*9 / (PPGR_meal_merged_filt['Carb(g)']*4+PPGR_meal_merged_filt['Protein(g)']*4+PPGR_meal_merged_filt['Fat(g)']*9) * 100

OTU_data = OTU_data.loc[:,list(np.sum(OTU_data>0) > len(OTU_data)*0.3)]
microbiome_col = list(OTU_data.columns)
OTU_data_bc = pd.DataFrame(stats.boxcox(OTU_data,lmbda=0.25),columns=OTU_data.columns,index=OTU_data.index)

# Features
CGM_pre30_col = [f"m_{t:03d}" for t in range(30,0,-5)]+['p_000']
futureCGM_120 = [f"p_{t:03d}" for t in range(5,121,5)]
meal_composition = ['Carb_root','Protein(g)','Fat(g)','Energy(kcal)','Fiber(g)','Carb_pro','Protein_pro','Fat_pro']
meal_context = ['meal_m1','prot_b6h','Time']
cli_col = ['Age','DM_Duration','BMI','HbA1c','SBP','DBP','HDL','LDL','AST(IU/L)','ALT(IU/L)'] 
med_col = ['Basal_Ins_Dose_Unit','MFM_Dose_Total','SU_Dose_Total','DPP4i_Dose_Total','SGLT2i_Dose_Total'] 


# Final dataframe
PPGR_data =  PPGR_meal_merged_filt.dropna(subset=['PPGR_u2','PPGR_u4']+CGM_pre30_col,axis=0)
full_data = pd.merge(PPGR_data[['ID','Dname','PPGR_u2','PPGR_u4','Carb(g)','Meal']+CGM_pre30_col+CGM_post240_col+meal_composition+meal_context],OTU_data_bc[microbiome_col],left_on='ID',right_index=True)
full_data = full_data.T.drop_duplicates().T
full_data = pd.merge(full_data,med_data,how='left',on='ID')
full_data = pd.merge(full_data,clinical_data,how='left',on='ID')
full_data['PPGR_u2'] = full_data['PPGR_u2'].astype(float)
full_data['PPGR_u4'] = full_data['PPGR_u4'].astype(float)

samples = list(full_data['ID'].unique())
full_data
```

```python
evaluation_data = pd.read_csv('../..//data/validation/PPGR_meal_merged_validation.csv',index_col=0)
evaluation_data =  evaluation_data.dropna(subset=CGM_pre30_col,axis=0)
evaluation_data
```

# Load optimized hyperparameter

```python
def extract_params(row):
    row = ast.literal_eval(row) 
    return {
        'microbiome_units': row['microbiome_units'],
        'n_steps': row['n_steps'],
        'dropout_rate': row['dropout_rate'],
        'learning_rate': row['learning_rate']
    }
DNN_LOOCV = pd.read_csv(f'DNN_{task}_LOOCV_multistep_summary.csv',index_col=0)
DNN_LOOCV['hyperparameter_bestparams'] = DNN_LOOCV['hyperparameter_bestparams'].apply(extract_params)
DNN_param = DNN_LOOCV['hyperparameter_bestparams'].value_counts().index[0]

XGB_LOOCV = pd.read_csv(f'XGB_{task}_LOOCV_summary.csv',index_col=0)
XGB_param = ast.literal_eval(XGB_LOOCV['hyperparameter_bestparams'].value_counts().index[0])

RF_LOOCV = pd.read_csv(f'RF_{task}_LOOCV_summary.csv',index_col=0)
RF_param = ast.literal_eval(RF_LOOCV['hyperparameter_bestparams'].value_counts().index[0])

SVM_LOOCV = pd.read_csv(f'SVM_{task}_LOOCV_summary.csv',index_col=0)
SVM_param = ast.literal_eval(SVM_LOOCV['hyperparameter_bestparams'].value_counts().index[0])
```

```python
tf.random.set_seed(2)
np.random.seed(2)
####### Original data ##########
x_train_cgm, x_train_meal_composition, x_train_meal_context,x_train_cli,x_train_med,x_train_microbiome = feature_split(full_data)
x_test_cgm, x_test_meal_composition, x_test_meal_context,x_test_cli,x_test_med,x_test_microbiome = feature_split(evaluation_data)
scaler = MinMaxScaler()
x_train_microbiome = scaler.fit_transform(x_train_microbiome)
x_test_microbiome = scaler.transform(x_test_microbiome)
if task =='2h':
    PPGR = 'PPGR_u2'
if task == '4h':
    PPGR = 'PPGR_u4'
train_ppgr = full_data[PPGR]
test_ppgr = evaluation_data[PPGR].astype(float)

model = DNN(DNN_param)
model.fit([x_train_cgm, x_train_meal_composition, x_train_meal_context,x_train_microbiome,x_train_cli,x_train_med],train_ppgr,
          epochs=15,batch_size=64,shuffle=False,verbose=0)
#model.save(f'./Optimized_models/All_features_{task}.h5')
DNN_predicted = model.predict([x_test_cgm, x_test_meal_composition, x_test_meal_context,x_test_microbiome,x_test_cli,x_test_med],verbose=0)

DNN_r, DNN_p = stats.pearsonr(test_ppgr, DNN_predicted.flatten())
DNN_RMSE = mse(test_ppgr, DNN_predicted.flatten()).numpy()**(1/2)
DNN_MAE = mae(test_ppgr, DNN_predicted.flatten()).numpy()

print('R ',DNN_r)
print('RMSE ',DNN_RMSE)
print('MAE ',DNN_MAE)
```

# Evaluation result visualization

```python
def Evaluation_vis(predicted,actual,model,task): 
    abs_error = np.abs(predicted-actual)
    guideline = np.linspace(-500,800,10)
    plt.figure(figsize=(5,3))
    if task == '2h':
        vmax=280
        plt.ylim(-50,200)
    else:
        vmax=650
        plt.ylim(-210,480)
    plt.scatter(y=predicted,x=actual,alpha=0.8,c=abs_error,cmap='coolwarm',s=7,vmin=0,vmax=vmax)
    plt.plot(guideline,guideline,'--',lw=1,color='gray')
    cbar = plt.colorbar()
    cbar.set_label('Absolute error (mg/dl•h)', rotation=90)
    res = stats.pearsonr(predicted, actual)
    plt.ylabel(f'Predicted PPGR ({task}, mg/dl•h)')
    plt.xlabel(f'Measured PPGR ({task}, mg/dl•h)')
    plt.xlim(np.min(actual)-30,np.max(actual)+30)

    ymin, ymax = plt.gca().get_ylim()
    plt.text(np.min(actual),ymin+(ymax-ymin)*9/10,f'R={np.round(res[0],3)}',fontsize=12)
    plt.text(np.min(actual),ymin+(ymax-ymin)*8.5/10,'p=%.2e'%res[1],fontsize=10)
    plt.title(model,fontsize=10)
    plt.savefig(f"./figure/{model}_prediction_{task}.jpeg",dpi=300,bbox_inches='tight',pad_inches=0.1)
    plt.show()
```

# Feature ablation

```python
tf.random.set_seed(2)
np.random.seed(2)

ablation_R_list,ablation_rmse_list,ablation_mae_list = [],[],[]
####### Original data ##########
x_train_cgm, x_train_meal_composition, x_train_meal_context,x_train_cli,x_train_med,x_train_microbiome = feature_split(full_data)
x_test_cgm, x_test_meal_composition, x_test_meal_context,x_test_cli,x_test_med,x_test_microbiome = feature_split(evaluation_data)
scaler = MinMaxScaler()
x_train_microbiome = scaler.fit_transform(x_train_microbiome)
x_test_microbiome = scaler.transform(x_test_microbiome)
if task =='2h':
    PPGR = 'PPGR_u2'
if task == '4h':
    PPGR = 'PPGR_u4'
train_ppgr = full_data[PPGR]
test_ppgr = evaluation_data[PPGR].astype(float)

for feature in [None,'CGM','Meal_composition','Meal_context','Clinical','Microbiome','Medication']:
    print(feature," Feature ablation")
    model = DNN_ablated(DNN_param,feature)
    model.fit([x_train_cgm, x_train_meal_composition, x_train_meal_context,x_train_microbiome,x_train_cli,x_train_med],train_ppgr,
              epochs=15,batch_size=64,shuffle=False,verbose=0)
    
    DNN_predicted = model.predict([x_test_cgm, x_test_meal_composition, x_test_meal_context,x_test_microbiome,x_test_cli,x_test_med],verbose=0)
    
    DNN_r, DNN_p = stats.pearsonr(test_ppgr, DNN_predicted.flatten())
    DNN_RMSE = mse(test_ppgr, DNN_predicted.flatten()).numpy()**(1/2)
    DNN_MAE = mae(test_ppgr, DNN_predicted.flatten()).numpy()

    ablation_R_list.append(DNN_r)
    ablation_rmse_list.append(DNN_RMSE)
    ablation_mae_list.append(DNN_MAE)
    
    print('R ',DNN_r)
    print('RMSE ',DNN_RMSE)
    print('MAE ',DNN_MAE)
    print('')
```

```python
fig,ax = plt.subplots(figsize=(4,3))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ablation_negative_rmse_list = [-RMSE for RMSE in ablation_rmse_list]
ablation_negative_mae_list = [-MAE for MAE in ablation_mae_list]
sct = ax.scatter(x=ablation_negative_rmse_list,y=ablation_negative_mae_list,s=100,c=ablation_R_list,cmap='RdYlBu_r',edgecolor='black')
if task=='2h':
    plt.xlim(-71,-56)
    plt.ylim(-55, -45.5)
    cax = ax.inset_axes([-60,-54.5,0.4,4],transform=ax.transData)
else:
    cax = ax.inset_axes([-124,-127,1.5,16],transform=ax.transData)
    plt.xlim(-170,-112)
    plt.ylim(-130,-91)
plt.xlabel('Negative RMSE')
plt.ylabel('Negative MAE')
plt.title(task)
#
cbar = plt.colorbar(sct, cax=cax,shrink=0.5)
cbar.set_label('Pearson\'s R', rotation=90)
plt.savefig(f'./figure/feature_ablation_{task}.jpeg',dpi=300)
plt.show()
```

# Subgroup prediction

```python
samples = list(evaluation_data['ID'].unique())
carb_sensitivity = []
carb_sensitivity_p = []

for i,sample in enumerate(samples):
    data = evaluation_data[evaluation_data['ID']==sample]
    r,p = stats.pearsonr(data['Carb(g)'],data['PPGR_u2'])
    carb_sensitivity.append(r)
    carb_sensitivity_p.append(p)
carb_sensitivity_df = pd.DataFrame({'ID':samples,'carb_sen':carb_sensitivity,'p':carb_sensitivity_p})
    
prediction_df = pd.DataFrame({'ID':evaluation_data['ID'],'actual':test_ppgr,'DNN_predicted':DNN_predicted.flatten(),'XGB_predicted':XGB_predicted.flatten(),
                              'RF_predicted':RF_predicted.flatten(),'SVM_predicted':SVM_predicted.flatten(),'carbo_predicted':carbo_predicted.flatten()})
prediction_df = pd.merge(prediction_df,carb_sensitivity_df,how='left',on='ID')
prediction_df.loc[prediction_df['p']<=0.05,'carb_sen_cate']='high responder'
prediction_df.loc[prediction_df['p']>0.05,'carb_sen_cate']='low responder'
prediction_df
```

```python
high = prediction_df[prediction_df['carb_sen_cate']=='high responder']
low = prediction_df[prediction_df['carb_sen_cate']=='low responder']

carb_high_rmse = mse(high['carbo_predicted'],high['actual']).numpy()**(1/2)
carb_low_rmse = mse(low['carbo_predicted'],low['actual']).numpy()**(1/2)
high_improvement = []
low_improvement = []
models=['RF','XGB','DNN','Average']
for model in models[:3]:
    model_high_rmse = mse(high[f'{model}_predicted'],high['actual']).numpy()**(1/2)
    model_low_rmse = mse(low[f'{model}_predicted'],low['actual']).numpy()**(1/2)
    high_improvement.append((carb_high_rmse-model_high_rmse)/carb_high_rmse*100)
    low_improvement.append((carb_low_rmse-model_low_rmse)/carb_low_rmse*100)

high_improvement.append(np.mean(high_improvement))
low_improvement.append(np.mean(low_improvement))
improvement_df = pd.DataFrame({'Subgroup':['High']*4+['Low']*4,'Improvement_rmse':high_improvement+low_improvement,'model':models+models})

colors =  ['#FDD2B1','#F18361','#CE6060','lightgrey']

fig,ax = plt.subplots(figsize=(1.5,3))
sns.scatterplot(data=improvement_df,x='Subgroup',y='Improvement_rmse',hue='model',style='model',s=120,palette=sns.color_palette(colors),edgecolor='black',alpha=0.8)
for i in range(len(models)):
    ax.plot([0,1],[high_improvement[i], low_improvement[i]], color=colors[i], ls='--', zorder=1,linewidth=1)
    
plt.ylabel('% Improvement of RMSE (vs. Carbo)')
plt.xlim(-0.5,1.5)
plt.ylim(0,25)
plt.legend(title=None,loc='upper center',bbox_to_anchor=(1.5, 0.4))
plt.title(task)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.savefig(f'./figure/subgroup_improvement_{task}.jpeg',dpi=300)
plt.show()
```

## Clustering

```python
def adj_mat_clustering(embedding, projection, n_clusters=10):
    adj_matrix = pairwise_distances(embedding,metric='manhattan')
    clustering = AgglomerativeClustering(n_clusters=n_clusters,metric='euclidean',linkage = 'ward').fit(embedding)#,linkage='complete',
    
    plt.figure(figsize=(6,4))
    plt.scatter(
        projection[:, 0],
        projection[:, 1],
        c=clustering.labels_,
        cmap=plt.cm.get_cmap('plasma', n_clusters),
        s=5,alpha=0.8,
        vmin = -0.5, vmax=n_clusters-0.5)
    plt.axis('off')
    
    cbar = plt.colorbar(norm=mcolors.Normalize(vmin=1, vmax=n_clusters),shrink=0.7)
    cbar.set_ticks(ticks=np.arange(0,n_clusters),labels=range(1,n_clusters+1))#
    cbar.set_label('Cluster', rotation=90)
    plt.xlim(-5.5,16.5)
    plt.gca().set_aspect('equal', 'datalim')
    plt.savefig(f'./figure/DNN_embedding_cluster_{task}.png',dpi=300,transparent=True)
    plt.show()

    return clustering.labels_

cluster_labels = adj_mat_clustering(embedding,projection,5)
```
